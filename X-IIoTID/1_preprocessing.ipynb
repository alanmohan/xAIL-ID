```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
from sklearn.metrics import matthews_corrcoef
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import time
import klib
```

```python
# Read the dataset
df = pd.read_csv('X-IIoTID dataset.csv')
```

### Data Pre-processing

```python
# Perform data cleaning
klib.data_cleaning(df)
```

```python
# Combine class1 and class2
df['combined'] = df['class1']+'-'+df['class2']
df.drop(['class1', 'class2'], axis=1, inplace=True)
df['combined'].value_counts()
```

```python
# Drop Timestamp, Date, Scr_IP, Des_IP as they are not useful
df = df.drop(columns = ['Timestamp', 'Date', 'Scr_IP', 'Des_IP'])
```

```python
# drop rows with missing values like 'None', '-', 'aza', '?', '#DIV/0!'
df = df[~df.isin(['None']).any(axis=1)]
df = df[~df.isin(['-']).any(axis=1)]
df = df[~df.isin(['aza']).any(axis=1)]
df = df[~df.isin(['?']).any(axis=1)]
df = df[~df.isin(['#DIV/0!']).any(axis=1)]
```

```python
df
```

```python
cols = df.columns
```

```python
# Ordinal Encode attack category
data_bin_attackcat = df[['combined']]
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories='auto')
oe.fit(data_bin_attackcat)
data_bin_attackcat = oe.transform(data_bin_attackcat)
```

```python
# in col anomaly_alert, replace FALSE with False and TRUE with True, to avoid error in model training
df['anomaly_alert'] = np.where(df['anomaly_alert'] == 'FALSE', 'False', df['anomaly_alert'])
df['anomaly_alert'] = np.where(df['anomaly_alert'] == 'TRUE', 'True', df['anomaly_alert'])
```

```python
# Split categorical and numerical columns
cols_one_hot = ['Protocol', 'Service', 'is_syn_only', 'Is_SYN_ACK', 'is_pure_ack', 'is_with_payload', 'FIN or RST','Bad_checksum','is_SYN_with_RST' ,'anomaly_alert','Login_attempt', 'Succesful_login', 'File_activity', 'Process_activity', 'read_write_physical.process', 'is_privileged']
cols_numeric = [i for i in cols if i not in cols_one_hot]
cols_cat = [i for i in cols if i in cols_one_hot]
print(cols_numeric)
print(cols_cat)
```

```python
data_bin = df.drop(columns = ['combined'], axis=1)
data_bin_hot = pd.get_dummies(data_bin, columns=cols_cat)
```

```python
attack_cat_df = pd.DataFrame(data_bin_attackcat, columns=['combined'])
data_bin_hot = data_bin_hot.reset_index(drop=True)
data_bin_hot2 = pd.concat([data_bin_hot,attack_cat_df],axis=1, join='inner',ignore_index=False)
```

```python
# one hot encode combined column
attack_cat_df = pd.DataFrame(data_bin_hot2['combined'], columns=['combined'])
data_bin_hot2 = pd.get_dummies(data_bin_hot2,columns=['combined'])
```

```python
cols_numeric.remove('class3')
cols_numeric.remove('combined')

# Convert numberic columns to float
data_bin_hot2[cols_numeric] = data_bin_hot2[cols_numeric].astype('float') 
# Normalize the data between 0 and 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_bin_hot2[cols_numeric] = scaler.fit_transform(data_bin_hot2[cols_numeric])
```

```python
# label encode class3 (attack label)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_bin_hot2['class3'] = le.fit_transform(data_bin_hot2['class3'])
```

```python
# save processed dataset
data_bin_hot2.to_csv('X-IIoTID dataset_processed.csv', index=False)
```
